```python
from parseResults import *
import numpy as np
```

```python
results = parse_file("ppResults_all.txt")
```

```python
def format_strings(results: [dict]):
    for (graph, data) in results:
        for (alg, subdict) in data.items():
            for (k,v) in subdict.items():
                if k=='#edges' or k=='weight':
                    subdict[k] = str(int(float(v)))
                elif k=='pp time' or k=='time':
                    fv = float(v)
                    if fv == 0.:
                        subdict[k] = str(int(fv))
                    elif fv < 0.01:
                        subdict[k] = "{:#.3e}".format(fv)
                    else:
                        subdict[k] = "{:#.4g}".format(fv)
format_strings(results)
```

```python
def test2(results: [dict], fname):
    header = ['#edges', '#vertices', 'pp time', 'time', 'weight']
    # cols = ['original', 'LC', '1-2', 'SD2']
    cols = ['1-2_LC', 'LC_1-2', 'SD2->1-2', '1-2->SD2']
    header_str1 = 'file, ' + ', '.join(np.repeat(header, len(cols))) + '\n'
    header_str2 = 'file, ' + ', '.join(cols * len(header)) + '\n'
    with open(fname + '.csv', 'w') as fp:
        fp.writelines([header_str1, header_str2])
        # for fname, data in results:
        for (graph, data) in results:
            line = graph + ', ' + \
                ', '.join([data[col][head] for head in header for col in cols]) + '\n'
            fp.write(line)


test2(results, 'output2/output')
```
